# Demographic

人口学特征统计，如性别、年龄。

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# 中文字体设置
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
data_dir = Path('../../data/preprocessed')
participants_file = data_dir / 'participants.csv'
```

```python
# 导入被试数据
demographic_data = pd.read_csv(participants_file)

# 描述统计
demographic_data.describe()
```

```python
# 性别分布
demographic_data['gender'].value_counts().plot(kind='bar')
plt.title('性别分布')
plt.xlabel('')

# x轴标签旋转
plt.xticks(rotation=0)

# 显示具体数值
for index, value in enumerate(demographic_data['gender'].value_counts()):
    plt.text(index, value, str(value), ha='center', va='bottom')
```

```python
# 年龄范围 (Mean ± SD)
age_range = (demographic_data['age'].min(), demographic_data['age'].max())
age_mean = demographic_data['age'].mean()
age_std = demographic_data['age'].std()

print(f'年龄范围: {age_range[0]} - {age_range[1]} ({age_mean:.2f} ± {age_std:.2f})')
```

```python
# 年龄直方图

# 计算以5为倍数的边界，确保覆盖所有数据
min_age = int(demographic_data['age'].min())
max_age = int(demographic_data['age'].max())

# 向下/向上取整到最近的5的倍数
start = (min_age // 5) * 5
end = (max_age // 5 + 1) * 5

# 创建以5岁为间隔的 bins
bins = range(start, end + 1, 5)

plt.figure(figsize=(8, 6))

# 绘制直方图
demographic_data['age'].plot(kind='hist', bins=bins, edgecolor='black', alpha=0.8)

plt.title('年龄分布')
plt.xlabel('年龄区间')
plt.ylabel('频数')

# 强制 x 轴刻度显示为 bins 的边界 (15, 20, 25...)
plt.xticks(bins)
plt.grid(axis='y', alpha=0.3, linestyle='--')

plt.show()
```
